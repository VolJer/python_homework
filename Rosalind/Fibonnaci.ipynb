#Fibonnaci rabbits

```python
def rabbits(n, k):
    if n == 1:
        return 1
    if n == 2:
        return 1
    if n >= 3:
        return rabbits(n-1,k) + rabbits(n-2,k)*k
```

```python
rabbits(n=34, k=4)
```

#Fibonnaci mortal rabbits

```python
def dead_rabbits(n, m):
    ages = [1] + [0] * (m - 1)
    
    for month in range(2, n + 1):
        new_borns = sum(ages[1:])
        ages = [new_borns] + ages[:-1]
    return sum(ages)
  
        
```
