You need a script that looks at your integers, takes the ones that are non-0, multiplies them by two for the correct offspring number and then multiplies each integer by the correct probability of dominant phenotype occurence.

```python
s= "16796 17314 19910 17704 18115 18076"
integers = [int(x) for x in s.split()]
probabilities = [1, 1, 1, 0.75, 0.5, 0]

sums=0
for i, count in enumerate(integers):
    expected = count*2*probabilities[i]
    sums+=expected
print(sums)
```
