# ORF

## Given: 
A DNA string s of length at most 1 kbp in FASTA format.

## Return: 
Every distinct candidate protein string that can be translated from ORFs of s. Strings can be returned in any order.


## Toy example:

s="AGCCATGTAGCTAACTCAGGTTACATGGGGATGACCCCGCGACTTGGATTAGAGTCTCTTTTGGAATAAGCCTGAATGATCCGAGTAGCATCTCAG"

## Example output:

MLLGSFRLIPKETLIQVAGSSPCNLS
M
MGMTPRLGLESLLE
MTPRLGLESLLE

```python
def rev_comp(s):
    comp = {'A':'T', 'T':'A', 'C':'G', 'G':'C'}
    return ''.join(comp[base] for base in s[::-1])
```

```python
def reading_frames(s):
    s_rna=s.replace("T","U")
    orf1=s_rna
    orf2=s_rna[1:]
    orf3=s_rna[2:]
    s_comp=rev_comp(s)
    rna_comp=s_comp.replace("T","U")
    orf4=rna_comp
    orf5=rna_comp[1:]
    orf6=rna_comp[2:]
    return[orf1, orf2, orf3, orf4, orf5, orf6]
```

```python
frames=reading_frames(s)
print(frames)
```

```python
def positions(orf):
    start=[]
    for i in range(0,len(orf),-2):
        codon=orf[i:i+3]
        if codon=="AUG":
            start.append(i)
    return start
```

```python
start=positions(orf)
start
```

```python
def RF_Translation(orf, start, codons):
    protein = ''
    for i in range(start, len(orf), 3):
        codon = orf[i:i+3]
        if len(codon)<3:
            break
        if codons[codon] is None:
            break
    
        protein += codons[codon]

    return protein
    
```

```python
def protein_finder(orf,codons):
    starts=positions(orf)
    proteins=[]
    for start in starts:
        protein=RF_Translation(orf, start, codons)
        if protein is not None:
            proteins.append(protein)
    return proteins
```

```python
protein_finder(frames[0], codons)
```

```python
def Full_ORF_function(s,codons):
    frames=reading_frames(s)
    proteins=[]
    for frame in frames:
        candidates=protein_finder(frame,codons)
        for protein in candidates:
            if protein not in proteins:
                proteins.extend([protein])
    print('\n'.join(proteins))
```
