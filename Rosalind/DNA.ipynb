```python
import utilities
fasta = utilities.read_file_into_lines('TrueGC.txt')
#fasta
#making a dictionary
def parse_fasta(lines_list):
    current_sequence = ''
    current_id = ''
    sequences = {}
    for line in lines_list:
        if line[0] == '>':
            sequences[current_id] = current_sequence
            current_id = line[1:]
            current_sequence = ''
        else:
            current_sequence += line

    sequences[current_id] = current_sequence
    del sequences['']
    return sequences

def GC_content(seq):
    count_C = seq.count("C")
    count_G = seq.count("G")
    GC_percent = (count_C + count_G)/len(seq) *100
    return(GC_percent)
    
sequences = parse_fasta(fasta)
#sequences
maximum_gc = -1
maximum_gc_id = ''

for seq_id, seq in sequences.items():
    gc = GC_content(seq)

    if gc > maximum_gc:

        maximum_gc = gc
        maximum_gc_id = seq_id

print(maximum_gc_id)
print(maximum_gc)
```
