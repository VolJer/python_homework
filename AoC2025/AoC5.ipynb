```python
import utilities
both=utilities.read_file_into_lines("inputAoC5.txt")
def gap_finder(l):
    for i, line in enumerate(l):
        if line == "":
            i=i
            return int(i)
```

```python
i =gap_finder(both)
print(i)
```

```python
ranges_base =both[0:i]
ranges_base

```

```python
ingredients_base=both[int(i+1):]
ingredients_base
```

```python
def range_checker(r,s):
    ranges=r
    fresh_ingredients=0
    for element in ranges:
        list_one= element.split("-")
        start=int(list_one[0])
        end=int(list_one[1])
        if s <= end and s>=start:
            fresh_ingredients+=1
            break
        else:
            continue
        
    return fresh_ingredients
```

```python
range_checker(ranges_base,5)
```

```python
ingredients=ingredients_base
fresh_ingredients=0
for ingredient in ingredients:
    ingredient=int(ingredient)
    fresh_ingredients += range_checker(ranges_base,ingredient)
    print(fresh_ingredients)
```
