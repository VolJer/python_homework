```python
def range_checker(r,s):
    ranges=r
    fresh_ingredients=0
    for element in ranges:
        list_one= element.split("-")
        start=int(list_one[0])
        end=int(list_one[1])
        if s <= end and s>=start:
            fresh_ingredients+=1
            break
        else:
            continue
        
    return fresh_ingredients
```

```python
data=["3-5",
"10-14",
"16-20",
"12-18",]
def range_edges(r):
    ranges=r
    start_list=[]
    end_list=[]
    for one_range in ranges:
        list_one= one_range.split("-")
        start=int(list_one[0])
        end=int(list_one[1])
        start_list.append(start)
        end_list.append(end)
    return start_list, end_list
```

write a function that JUST checks for overlaps.

```python
def overlapper(a,b):
    range_a=a.split("-")
    range_b=b.split("-")
    i=int(range_a[0])
    j=int(range_a[1])
    k=int(range_b[0])
    l=int(range_b[1])
    if i <= k and k <= j or i>=k and i <= l:
        return True
    elif k<=i and l>=j or k>=i and l<=j:
        return True
    else:
        return False
```

```python
overlapper("12-18", "16-20")
```

```python
def merger(a, b):
    range_a=a.split("-")
    range_b=b.split("-")
    i=int(range_a[0])
    j=int(range_a[1])
    k=int(range_b[0])
    l=int(range_b[1])
    start_list=[i,k]
    end_list=[j,l]
    merged_range=[]
    merged_range=[str(min(start_list)), str(max(end_list))]
    merged="-".join(merged_range)
    return merged
```

```python
merger("12-18", "16-20")
```

```python
data=["3-5",
"10-14",
"16-20",
"12-18",]
merged_ranges=merger("12-18","16-20")
data.append(merged_ranges)
print(data)
```
