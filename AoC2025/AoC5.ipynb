```python
def range_checker(s):
    ranges=["3-5","10-14","16-20","12-18"]
    fresh_ingredients=0
    for element in ranges:
        list_one= element.split("-")
        start=int(list_one[0])
        end=int(list_one[1])
        if s <= end and s>=start:
            fresh_ingredients+=1
        else:
            continue
        
    return fresh_ingredients
```

```python
range_checker(5)
```

```python
ingredients= [1,5,8,11,17,32]
for ingredient in ingredients:
    fresh_ingredients = range_checker(ingredient)
    print(fresh_ingredients)
```
