```python
def read_file_into_lines(file_path):
    lines = []
    with open(file_path, 'r') as infile:
        for line in infile.readlines():
            cleaned = line.rstrip()
            lines.append(cleaned)
    return lines
```

```python
data= read_file_into_lines("/home/jvegsb01/python_homework/AoC2025/AoC4.txt")
```

```python
coordinates=[]
for lines in range(0, len(data)):
    line=data[lines]

    for paper in range(len(line)):
        position= paper
        coordinate= [lines, position]
        coordinates.append(coordinate)


```

```python

def adjacents(x,y):
    adj1=[x-1,y]
    adj2=[x+1,y]#line is fine
    adj3=[x-1,y-1]
    adj4=[x,y-1]
    adj5=[x+1,y-1]
    adj6=[x-1,y+1]
    adj7=[x,y+1]
    adj8=[x+1,y+1]
    
    list_of_adjacents=[adj1,adj2,adj3,adj4,adj5,adj6,adj7,adj8]

    return list_of_adjacents
```

```python
available_paper=0
for lines in range(0, len(data)):
    line=data[lines]

    for paper in range(len(line)):
        position=line[paper]
        if position == '@':
            x = lines
            y = paper
            list_of_adjacents=(adjacents(x,y))
            can_counter=0
            for value in list_of_adjacents:
                x_val=value[0]
                y_val=value[1]
                if  x_val < 0:
                    continue
                if x_val >= len(data):
                    continue
                if y_val < 0:
                    continue
                if y_val >= len(data[x_val]):
                    continue
                if data[x_val][y_val] == '@':
                    can_counter += 1
            if can_counter <4:
                available_paper+=1

print(available_paper)
```
