```python
def zero_crosser(dial, r):
    direction = r[0]
    steps = int(r[1:])

    if direction == 'L':
        new_dial = dial - steps
        zero_count = new_dial // -100
        zero_count = max(zero_count, 0)
    else:
        new_dial = dial + steps
        zero_count = new_dial // 100

    if new_dial < 100:
        at_zero = new_dial % 100 == 0
        passed_zero = new_dial * dial < 0
        if at_zero or passed_zero:
            zero_count += 1

    return new_dial % 100, zero_count
```
