```python

def dial_turner(start,rotation):
    direction = rotation[0]
    steps=int(rotation[1:])
    if direction == "L":
        start= start - steps
    else:
        start= start + steps

    if start >= 100 or start < 0:
        start = start%100
    else:
        start=start
    
    return start
        
```

```python
import utilities
start=50
rotations= utilities.read_file_into_lines("/home/jvegsb01/python_homework/AoC2025/inputAoC.txt")
zero_count= 0
for rotation in rotations:
    start = dial_turner(start,rotation)
    if start == 0:
        zero_count+=1
print(zero_count)
```
